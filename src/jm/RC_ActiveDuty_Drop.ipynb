```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Admin\Documents\GitHub\Horse\data_preprocessing\RC_ActiveDuty.csv')

############################
#결측치 처리
df = df[df['noracefl'] == '정상']
############################


############################
#컬럼 드랍
drop_cols = [
    'chaksun', 'diffTot', 'rcSex', 'noracefl',
    'rcFrflag', 'rcHrfor', 'rcAge',
    'prow', 'prowName',
    'rcDiff2', 'rcDiff3', 'rcDiff4', 'rcDiff5',
    'rcSpcbu',
    'test_rank', 'rank'
]
df = df.drop(columns=drop_cols, errors='ignore')
############################


df.to_csv(r'race_cleaned.csv', index=False, encoding='utf-8-sig')
```
