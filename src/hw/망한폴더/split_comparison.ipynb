```python
from pathlib import Path
import json
import pandas as pd
from IPython.display import display
```

```python
ref_path = Path('split_comparison_reference.json')
ref = json.loads(ref_path.read_text(encoding='utf-8'))
rows = []
for split_name, split_obj in ref.items():
    for model_name, metrics in split_obj.items():
        if model_name == 'description':
            continue
        row = {'split': split_name, 'model': model_name}
        row.update(metrics)
        rows.append(row)
df = pd.DataFrame(rows)
display(df)
```

```python
pivot_rmse = df.pivot(index='model', columns='split', values='rmse')
pivot_win = df.pivot(index='model', columns='split', values='winner_accuracy')
pivot_top3 = df.pivot(index='model', columns='split', values='winner_in_top_3_rate')
display(pivot_rmse)
display(pivot_win)
display(pivot_top3)
```
