```python
df = df.copy()

# 날짜 변환
df["경주일자"] = pd.to_datetime(df["경주일자"])

# 정렬 (매우 중요)
df = df.sort_values(["마명", "경주일자"]).reset_index(drop=True)

# 순위비율 (핵심)
df["순위비율"] = df["순위"] / df["출주두수"]

# 방향 뒤집기 (높을수록 좋게)
df["폼점수"] = 1 - df["순위비율"]
```

```python
df["폼_3"] = (
    df.groupby("마명")["폼점수"]
    .rolling(3)
    .mean()
    .reset_index(level=0, drop=True)
)
```

```python
df["폼_5"] = (
    df.groupby("마명")["폼점수"]
    .rolling(5)
    .mean()
    .reset_index(level=0, drop=True)
)
```

```python
def rolling_30d(group):
    return (
        group
        .set_index("경주일자")["폼점수"]
        .rolling("30D")
        .mean()
    )

df["폼_30일"] = (
    df.groupby("마명", group_keys=False)
      .apply(rolling_30d)
      .reset_index(drop=True)
)
```

```python
cols = ["폼_3", "폼_5", "폼_30일"]

for col in cols:
    df[col] = df.groupby("마명")[col].shift(1)
```

```python
df["직전폼"] = df.groupby("마명")["폼점수"].shift(1)
```

```python
df["폼변화"] = df["폼_3"] - df["폼_5"]
```

```python
df["휴식일"] = (
    df.groupby("마명")["경주일자"]
    .diff()
    .dt.days
)
```

```python
df["거리폼"] = (
    df.groupby(["마명", "경주거리"])["폼점수"]
    .transform("mean")
)
```

```python
df["최고폼"] = (
    df.groupby("마명")["폼점수"]
    .transform("max")
)
```

```python
df["입상"] = (df["순위"] <= 3).astype(int)
```

```python
df[["마명", "폼_3", "폼_5", "폼_30일"]].sort_values("폼_3").head(10)
```

```python
features = [
    "폼_3", "폼_5", "폼_30일",
    "직전폼", "폼변화", "휴식일",
    "거리폼", "최고폼"
]

train_df = df.dropna(subset=features + ["입상"])

X = train_df[features]
y = train_df["입상"]
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, shuffle=False
)

model = LGBMClassifier(n_estimators=300)
model.fit(X_train, y_train)

pred = model.predict_proba(X_test)[:,1]

print("AUC:", roc_auc_score(y_test, pred))
```

```python
import pandas as pd

importance = pd.DataFrame({
    "feature": features,
    "importance": model.feature_importances_
}).sort_values("importance", ascending=False)

print(importance)
```
