```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display

from analyze_overfitting_advanced import run_overfitting_analysis
```

```python
summary = run_overfitting_analysis()
display(pd.DataFrame([summary]))
```

```python
curve_df = pd.read_csv(summary['artifacts']['curve_csv'], encoding='utf-8-sig')
display(curve_df.head())
display(curve_df.tail())
```

```python
display(Image(filename=summary['artifacts']['curve_png']))
```

```python
plt.figure(figsize=(10,6))
plt.plot(curve_df['iteration'], curve_df['train_rmse'], label='Train RMSE', linewidth=2)
plt.plot(curve_df['iteration'], curve_df['test_rmse'], label='Test RMSE', linewidth=2)
best_iter = int(summary['best_test_iteration'])
best_test = float(summary['best_test_rmse'])
plt.axvline(best_iter, color='gray', linestyle='--', linewidth=1.5, label='Best test iteration')
plt.scatter([best_iter], [best_test], color='red', zorder=5)
plt.title('Advanced Model Overfitting Check')
plt.xlabel('Boosting Iteration')
plt.ylabel('RMSE')
plt.legend()
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```
