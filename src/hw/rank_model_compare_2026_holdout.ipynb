```python
from pathlib import Path
import json
import pandas as pd
from IPython.display import display
```

```python
files = [
    ('baseline', Path('rank_model_results.json')),
    ('history_features', Path('rank_model_history_results.json')),
    ('advanced_features', Path('rank_model_advanced_results.json')),
]
rows = []
for label, path in files:
    obj = json.loads(path.read_text(encoding='utf-8'))
    row = {'label': label}
    row.update(obj['test_metrics'])
    row.update({
        'train_rows': obj['split_rows']['train'],
        'valid_rows': obj['split_rows']['valid'],
        'test_rows': obj['split_rows']['test'],
        'train_before': obj['split_rule']['train_before'],
    })
    rows.append(row)
compare = pd.DataFrame(rows)[['label','mae','rmse','r2','winner_accuracy','winner_in_top_3_rate','ndcg_at_3','train_rows','valid_rows','test_rows','train_before']]
display(compare.sort_values('rmse'))
```

```python
for label, path in files:
    obj = json.loads(path.read_text(encoding='utf-8'))
    print('---', label, '---')
    print(json.dumps(obj['selected_lightgbm_params'], ensure_ascii=False, indent=2))
```
