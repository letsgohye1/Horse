##### 날씨 '-' 인경우 처리

```python
import pandas as pd

# 1. '-'를 NaN으로 변경
df['weath'] = df['weath'].replace('-', pd.NA)

# 2. 같은 날짜 기준으로 가장 많이 나온 날씨로 채우기
df['weath'] = df.groupby('rcDate')['weath'].transform(
    lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else pd.NA)
)
```

```python
import pandas as pd

# 1. 경주 단위 정의
race_id = ['rcDate', 'rcNo']

# -----------------------------
# 2. 경주별 말 개수 확인
# -----------------------------
race_size = df.groupby(race_id).size()

print("=== 경주당 말 개수 통계 ===")
print(race_size.describe())

print("\n=== 말 수 이상한 경주 (하위) ===")
print(race_size.sort_values().head(10))

# -----------------------------
# 3. rcTime 분산 확인
# -----------------------------
time_nunique = df.groupby(race_id)['rcTime'].nunique()

print("\n=== rcTime 다양성 분포 ===")
print(time_nunique.value_counts().sort_index())

# -----------------------------
# 4. rcTime으로 rank 생성 테스트
# -----------------------------
df['test_rank'] = df.groupby(race_id)['rcTime'].rank(method='min')

rank_check = df.groupby(race_id)['test_rank'].nunique()

print("\n=== rank 중복 분포 ===")
print(rank_check.value_counts().sort_index())

# -----------------------------
# 5. 경주 내 rcTime 단조성 체크 (정상 순위 여부)
# -----------------------------
monotonic_check = df.groupby(race_id)['rcTime'].apply(lambda x: x.is_monotonic_increasing)

print("\n=== rcTime 정렬성 비율 ===")
print(monotonic_check.value_counts(normalize=True))

# -----------------------------
# 6. 최종 판단
# -----------------------------
valid_ratio = (time_nunique > 1).mean()

print("\n=== 최종 판단 ===")
print("rank 생성 가능 비율:", valid_ratio)
```
