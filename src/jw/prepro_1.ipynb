```python
import pandas as pd

df = pd.read_csv("../../data_raw/raw_race_2023_to_20260426.csv", encoding="utf-8-sig")

df.head()
```

```python
print(f"df 데이터 개수: {df.shape[0]:,}")
```

```python
def show_missing(df: pd.DataFrame):
    """df의 결측 항목들 확인"""
    null_counts = df.isnull().sum()
    null_cols = null_counts[null_counts > 0]
    print(null_cols)


show_missing(df)
```

```python
# 제거할 결측 항목
target_columns = [
    "prow",
    "prowName",
    "prtr",
    "prtrName",
    "rcDiff2",
    "rcDiff3",
    "rcDiff4",
    "rcDiff5",
    "rcP1Odd",
    "rcP1Sale",
    "rcP2Odd",
    "rcP2Sale",
    "rcP3Odd",
    "rcP3Sale",
    "rcP4Odd",
    "rcP4Sale",
    "rcP5Odd",
    "rcP5Sale",
    "rcP6Odd",
    "rcP6Sale",
    "rcP8Odd",
    "rcP8Sale",
    "track",
]
df = df.dropna(subset=target_columns)

show_missing(df)
```

```python
# csv 파일 저장
df.to_csv("../../data_preprocessing/cleaned_data_1.csv", index=False)
```

```python
df2 = pd.read_csv("../../data_preprocessing/cleaned_data_1.csv", encoding="utf-8-sig")

print(f"df2 데이터 개수: {df2.shape[0]:,}")
show_missing(df2)
```
